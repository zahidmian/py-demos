```python
import numpy as np
import pandas as pd
from numpy.random import randn

#stats
from scipy import stats

#plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
dat1 = randn(100)
dat2 = randn(100)
```

```python
sns.boxplot(data=[dat1,dat2])
```

```python
sns.boxplot(data=[dat1,dat2], whis=np.inf)
```

```python
data1 = stats.norm(0,5).rvs(100)

# two gamm dist 
data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                      -1* stats.gamma(5).rvs(50)])

sns.boxplot(data=[data1, data2], whis=np.inf)
```

```python
sns.violinplot(data=[data1,data2])
```

```python
sns.violinplot(data=[data1], innner="quartile")
```
